```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('books.csv')

months_order = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
                'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
df['Месяц'] = pd.Categorical(df['Месяц'], categories=months_order, ordered=True)

df = df.sort_values(['Год', 'Месяц'])

df_2023 = df[df['Год'] == 2023]
df_2024 = df[df['Год'] == 2024]

def plot_books_2023():
    """Линейный график книг за 2023 год"""
    plt.plot(df_2023['Месяц'], df_2023['Книги'],
             marker='o',
             color='red',
             linewidth=2,
             label='Прочитано книг')
    plt.title('Прочитанные книги 2023')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество книг')
    plt.grid(True, alpha=0.3)
    plt.legend()

    for x, y in zip(df_2023['Месяц'], df_2023['Книги']):
        plt.text(x, y, f'{y}', ha='center', va='bottom', fontsize=9, fontweight='bold')

def plot_pages_2024():
    """Столбчатая диаграмма страниц за 2024 год"""
    plt.bar(df_2024['Месяц'], df_2024['Страницы'],
            color='green',
            alpha=0.7,
            label='Прочитано страниц')
    plt.title('Прочитанные страницы 2024')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество страниц')
    plt.grid(True, alpha=0.3)
    plt.legend()

    for x, y in zip(df_2024['Месяц'], df_2024['Страницы']):
        plt.text(x, y, f'{y}', ha='center', va='bottom', fontsize=9, fontweight='bold')


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plot_books_2023()

plt.subplot(1, 2, 2)
plot_pages_2024()


plt.tight_layout()


plt.show()


print("=== ИНФОРМАЦИЯ О ДАННЫХ ===")
print(f"Всего записей: {len(df)}")
print(f"Данные за 2023 год: {len(df_2023)} месяцев")
print(f"Данные за 2024 год: {len(df_2024)} месяцев")
print("\nЗагруженные данные:")
print(df.to_string(index=False))
```
